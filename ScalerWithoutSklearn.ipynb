```python
import pandas as pd
import numpy as np
```

```python
arr = np.array([['M',81.4,84.0,44,6.2,120000,'no'],
              ['F',75.2,86.2,30,6.1,67000,'no'],
              ['F',80.0,86.7,29,5.9,45000,'yes'],
              ['F',89.8,72.2,60,5.6,50000,'yes'],
              ['M',88.3,90.3,46,7.2,80000,'no'],
              ['F',67.4,80.6,24,5.11,70000,'yes']])
```

```python
df = pd.DataFrame(arr)
```

```python
df
```

```python
df.columns = ['gender','p1','p2','age','height','salary','condition']
```

```python
df
```

```python
df[['age','salary']] = df[['age','salary']].astype(int)
df[['p1','p2','height']] = df[['p1','p2','height']].astype(float)
```

```python
def normal(x):
    return (x - x.min())/(x.max()-x.min())               #Function for min max
```

```python
cols = ['p1','p2','age','height','salary']

#calling the function
df[cols] = df[cols].apply(normal)
```

```python
df   #normalized dataframe
```

```python
def standard(x):
    return (x - x.mean())/x.std()
```

```python
cols = ['p1','p2','age','height','salary']

#calling the function
df[cols] = df[cols].apply(standard)
```
